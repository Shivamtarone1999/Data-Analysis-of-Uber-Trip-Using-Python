```python
%pylab inline
import pandas
import seaborn
```

```python
data = pandas.read_csv('Desktop/uber-raw-data-apr14.txt')
```

```python
data.tail()
```

# Convert dateTime and add some useful colomns

```python
data['Date/Time'] = data['Date/Time'].map(pandas.to_datetime)
```

```python
data.head()
```

```python
def get_dom(dt):
    return dt.day
data['dom'] = data['Date/Time'].map(get_dom)
```

```python
data.tail()
```

```python
def get_weekday(dt):
    return dt.weekday()

data['weekday'] = data['Date/Time'].map(get_weekday)

def get_hour(dt):
    return dt.hour

data['hour'] = data['Date/Time'].map(get_hour)

data.tail()
```


# Analysis

```python
hist(data.dom,bins = 30,rwidth =.8,range=(0.5,30.5))
xlabel('date of month')
ylabel('frequncy')
title('Frequency by date of month - UBER - Aprl-2019')
```

```python
#for k , rows in data.groupby('dom'):
#    print((k,len(rows)))

def count_rows(rows):
    return len(rows)

by_date = data.groupby('dom').apply(count_rows)
by_date
```

```python
plot(by_date)
```

```python
bar(range(1,31),by_date)
```

```python
by_date_sorted = by_date.sort_values()
by_date_sorted
```

```python
bar(range(1,31),by_date_sorted)
xticks(range(1,31),by_date_sorted.index)
xlabel('Date of month')
ylabel('Frequency')
title('Frequency of Date of Month - UBER -  April- 2019 ')
;
```
